```python
nd=pd.read_csv('finaltrain.csv')
```

```python
df_encoded = pd.get_dummies(nd, columns=['country'], prefix=['country'])
df_encoded = pd.get_dummies(nd, columns=['FpCOT'], prefix=['FpCOT'])
```

```python
df_encoded.to_csv('encoded_data.csv', index=False)
from sklearn.preprocessing import MinMaxScaler
df_encoded.head()
```

```python
scaler=MinMaxScaler()
cols=['credit_card_number']
dat=df_encoded[cols]
scaler.fit(dat)
newdat=scaler.transform(dat)
df_encoded[cols]=newdat
df_encoded.drop('Unnamed: 0',axis=1,inplace=True)
df_encoded.head()
```

```python
mean=df_encoded['credit_card_number'].mean()
```

```python
df_encoded['credit_card_number'].fillna(mean,inplace=True)
```

```python
df_encoded['cvv'].fillna(mean,inplace=True)
```

```python
x=df_encoded['state']
y=df_encoded.drop('state',axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=50)
```

```python
model.fit(y,x)
```
